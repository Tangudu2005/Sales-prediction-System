```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/Phone_Order_details.csv")
print(data.head())
```

```python
print(data.columns)
data.info()
```

```python
print(data.isnull().sum())
```

```python
product_sales = data['Product'].value_counts()
print(product_sales)
```

```python
product_sales.plot(kind='bar')

plt.title("Mobile Phone Sales Analysis")
plt.xlabel("Phone Model")
plt.ylabel("Number of Orders")

plt.show()
```

```python
product_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title("Phone Sales Distribution")
plt.ylabel("")

plt.show()
```

```python
most_sold = data['Product'].value_counts().idxmax()
print("Most Sold Phone:", most_sold)
```
